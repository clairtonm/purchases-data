```python
departments = {"sport": ['surf_board','skate', 'football_shoes','golf_shoes','soccer_shoes','running_shoes','soccer_ball','football_shoes','racket','bicycle','bicycle_helmet','sport_gloves','tennis_ball','golf_ball'], "tech": ["ram_memory","ssd","computer","laptop","monitor","mouse	","keyboard","usb_cable","wifi_adapter","hdmi_cable","headset","laptop_case","usb_stick"], "Clothing": ["basic_tshirt","basic_jeans","basic_dress","slim_jeans","shoes","jacket","neck_tshirt","polo_tshirt","sleeve_tshirt","sleeveless_tshirt","short","jeans_skirt","pants","shirt","heels","socks"], "Books": ["the_great_gatsby","the_grapes_of_wrath","nineteen_eighty_four","to_kill_a_mockingbird","the_lord_of_the_rings","brave_new_world","pride_and_prejudice","animal_farm","wuthering_heights","the_chronicles_of_narnia","moby_dick","the_hobbit","don_quixote","american_gods","neverwhere","the_ocean_at_the_end_of_the_lake"]}
```

```python
departments['sport']
```

```python
clients = raw_df['first_name'].tolist()
```

```python
df_purchases = pd.DataFrame(columns=['client','item'])
```

```python
client_purchases = np.empty_like([[1,2]])
```

```python
for client in clients :
    
    # Get the items from two departments
    department1 = random.choice(list(departments.keys()))
    department2 = random.choice(list(departments.keys()))
    print(department1, department2)
    
    while department1 == department2 :
        department2 = random.choice(list(departments.keys()))
        print(department2)
        
    items = []
    items1 = departments[department1]
    items2 = departments[department2]
    items.append(items1)
    items.append(items2)
    print(items1)
    print(items1)
    print(items)
    
    purchases = random.sample(items, random.randint(3,6))
    
    for purchase in purchases:
        print(client, purchase)
        client_purchases = np.append(client_purchases, [[client, purchase]], axis=0)
    
    
```

```python
client_purchases_list = client_purchases.tolist()
```

```python
client_purchases_list.pop(0)
```

```python
purchases = np.array(client_purchases_list)
```

```python
df_purchases = pd.DataFrame(purchases, columns=['client', 'item'])
```
